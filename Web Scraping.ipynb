```python
# using the try and exception method (because if the url is wrong then it can lead to the crash in the programe)
# if there is any mistake it will show the mistake that is happen.
try:
    response = requests.get("https://www.estesparkweather.net/archive_reports.php?date=202005") 
    # getting the url data by using the request command and storing in the response variable. 
    # In this command all the html contents are stored in the response varible.
    soup = BeautifulSoup(response.text,"html.parser") 
    # this command used to check the html tags that in the web site.
    # html parser is used to parse the html content.
    # parse is used to divide the html contents
    print (soup)
except Exception as e: # for exception purpose
    print (e)
```
